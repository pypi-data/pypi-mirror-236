# Booth Function Tests (MIN)
* $f(x,y) = (x+2y-7)^2 + (2x+y-5)^2$
* $[-10 \leq x,y \leq 10 ]$
* ## Optimal solution: [1,3],
* ## value: 0

```python
# get the things we need set up

# ** GA_Agent manager **
gaManager = GA_OptimizerOmegaMin(
         chromosome_size=2, 
         population_size=100, 
         chromosomeAlphabet=list(range(-10, 11)), 
         generations=100,
         mutation_rate=.05,
         crossOver_rate=.90,
         objective_func=boothFunction,
         mode="min",
        )
# ** GA_Agent Environment **

gaENV = GA_ENV(agentManager=gaManager, generations=100, patienceLimit=np.inf)



# ** Solve for optimal solution**
gaENV.Solve(verbose=True, interval=50)


# ** plot the progress toward the optimal solution by generation and show what it was and the score:


gaENV.getBest()

gaENV.plot_scores(title="Booth Function Peformance (MIN)")
```

# Matyas Function Tests (MIN)
* $0.26(x^2 + y^2) - 0.48xy$
* Range: $-10 \leq x,y \leq 10$
* ## Optimal solution: [0,0],
* ## value: 0

```python
def matyas(XY):
    X = XY[0]
    Y = XY[1]
    return .26*(X**2 + Y**2) - .48*X*Y

matyas([0,0])
```

```python
# get the things we need set up

# ** GA_Agent manager **
gaManager = GA_OptimizerOmegaMin(
         chromosome_size=2, 
         population_size=100, 
         chromosomeAlphabet=list(range(-10, 11)), 
         generations=100,
         mutation_rate=.05,
         crossOver_rate=.90,
         objective_func=matyas,
         # mode="min",
         mode="max",
        )
# ** GA_Agent Environment **

gaENV = GA_ENV(agentManager=gaManager, generations=100, patienceLimit=np.inf)



# ** Solve for optimal solution**
gaENV.Solve(verbose=True, interval=50)


# ** plot the progress toward the optimal solution by generation and show what it was and the score:


gaENV.getBest()

gaENV.plot_scores(title="Matyas Function Peformance (MIN)")
```

# Levi Function N. 13 Tests (MIN)
* $f(xy) = sin^2 3 \pi x + (x-1)^2 (1 + sin^2 3 \pi y) + (y-1)^2(1+sin^22\pi y)$
* Range: $-10 \leq x,y \leq 10$
* ## Optimal solution: [1,1],
* ## value: 0

```python
def radsFromDeg(deg):
    return deg * ((2*np.pi)/360)

def LeviN13(XY):
    X = XY[0]
    Y = XY[1]
    return np.around(np.sin(3*np.pi*X)**2 + ((X-1)**2)*(1+np.sin(3*np.pi*Y)**2) + ((Y-1)**2)*(1 + np.sin(2*np.pi*Y)**2 ), 4)

LeviN13([1,1])
```

```python
# get the things we need set up

# ** GA_Agent manager **
gaManager = GA_OptimizerOmegaMin(
         chromosome_size=2, 
         population_size=100, 
         chromosomeAlphabet=list(range(-10, 11)), 
         generations=100,
         mutation_rate=.05,
         crossOver_rate=.90,
         objective_func=LeviN13,
         mode="min",
         # mode="max",
        )
# ** GA_Agent Environment **

gaENV = GA_ENV(agentManager=gaManager, generations=100, patienceLimit=np.inf)



# ** Solve for optimal solution**
gaENV.Solve(verbose=True, interval=50)


# ** plot the progress toward the optimal solution by generation and show what it was and the score:


gaENV.getBest()

gaENV.plot_scores(title="LeviN13 Function Peformance (MIN)")
```
